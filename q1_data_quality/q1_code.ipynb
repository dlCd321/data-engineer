```python
# 统一派生字段：后续所有统计都基于这个副本，不改动原始 df。
work = df.assign(
    InvoiceStr=df["Invoice"].astype(str),
    StockCodeStr=df["StockCode"].astype(str).str.strip(),
    InvoiceDateParsed=pd.to_datetime(df["InvoiceDate"]),
    LineAmount=df["Quantity"] * df["Price"],
)
work = work.assign(
    InvoicePrefix=work["InvoiceStr"].str.extract(r"^([A-Za-z]+)", expand=False).fillna("numeric"),
    IsCancellation=work["InvoiceStr"].str.startswith("C", na=False),
    IsBadDebtAdjustment=work["InvoiceStr"].str.startswith("A", na=False),
    IsCustomerMissing=work["Customer ID"].isna(),
    IsDescriptionMissing=work["Description"].isna(),
)
work = work.assign(
    IsDamageOrInventoryLoss=(
        work["Price"].eq(0)
        & work["Quantity"].lt(0)
        & ~work["IsCancellation"]
        & ~work["IsBadDebtAdjustment"]
    )
)

def pct(value, denominator):
    return round(value / denominator * 100, 2) if denominator else np.nan

total_rows = len(work)
total_amount = work["LineAmount"].sum()
print(f"rows={total_rows:,}, net line amount={total_amount:,.2f}")
```

### 1. Invoice 前缀是交易类型

`C` 是取消/退货；`A` 是 bad debt adjustment。这里需要先定义销售口径：

- **Gross product sales**：通常只统计正向商品销售，排除 `C` 取消单、`A` bad debt、费用/邮费/折扣等非商品行。
- **Net product sales**：可包含 `C` 取消/退货来抵减商品销售，但仍应把 `A` bad debt 和费用类 StockCode 单独列示。
- **财务总额**：可以保留费用、邮费、佣金、bad debt，但要单独分组，否则会污染商品销量和客单价分析。

本数据中 `C` 有 **19,494 行**，金额合计 **-1,526,667.86**；`A` 只有 **6 行**，都是 `Adjust bad debt`，其中 **5 行 Price 为负数**。

```python
invoice_prefix_summary = (
    work.groupby("InvoicePrefix", dropna=False)
    .agg(
        rows=("Invoice", "size"),
        quantity_sum=("Quantity", "sum"),
        amount_sum=("LineAmount", "sum"),
        negative_price_rows=("Price", lambda s: (s < 0).sum()),
        zero_price_rows=("Price", lambda s: (s == 0).sum()),
        missing_customer_rows=("Customer ID", lambda s: s.isna().sum()),
    )
    .sort_values("rows", ascending=False)
)
display(invoice_prefix_summary)

display(
    work.loc[
        work["IsBadDebtAdjustment"],
        ["Invoice", "StockCode", "Description", "Quantity", "InvoiceDate", "Price", "Customer ID", "Country", "LineAmount"],
    ]
)
```

```python
import re


def classify_stock_code(code_value):
    code_text = str(code_value).strip().upper()
    if code_text in {"POST", "DOT", "C2"}:
        return "postage_carriage_dotcom"
    if code_text == "D":
        return "discount"
    if code_text in {"M", "ADJUST", "ADJUST2"}:
        return "manual_adjustment"
    if code_text in {"BANK CHARGES", "AMAZONFEE", "CRUK", "B"}:
        return "fees_commission_bad_debt"
    if code_text.startswith("GIFT"):
        return "gift_voucher"
    if code_text in {"S", "TEST001", "TEST002", "PADS"} or code_text.startswith("DCGS"):
        return "sample_test_internal"
    if re.fullmatch(r"\d+[A-Z]*", code_text):
        return "product_like"
    return "other_non_product_or_unknown"

work = work.assign(StockCodeCategory=work["StockCodeStr"].map(classify_stock_code))

stock_category_summary = (
    work.groupby("StockCodeCategory")
    .agg(
        rows=("StockCode", "size"),
        distinct_codes=("StockCodeStr", "nunique"),
        amount_sum=("LineAmount", "sum"),
    )
    .assign(
        row_pct=lambda x: (x["rows"] / total_rows * 100).round(3),
        amount_pct=lambda x: (x["amount_sum"] / total_amount * 100).round(3),
    )
    .sort_values("rows", ascending=False)
)
display(stock_category_summary)

non_product_code_list = (
    work.loc[~work["StockCodeCategory"].eq("product_like")]
    .groupby("StockCodeCategory")["StockCodeStr"]
    .agg(lambda s: ", ".join(sorted(s.astype(str).unique())))
    .reset_index(name="codes")
)
display(non_product_code_list)

work = work.assign(
    IsGrossProductSale=(
        work["InvoicePrefix"].eq("numeric")
        & work["StockCodeCategory"].eq("product_like")
        & work["Quantity"].gt(0)
        & work["Price"].gt(0)
    ),
    IsNetProductLine=(
        work["StockCodeCategory"].eq("product_like")
        & ~work["IsBadDebtAdjustment"]
        & work["Price"].gt(0)
        & work["Quantity"].ne(0)
    ),
)

sales_measure_summary = pd.DataFrame(
    [
        {"measure": "gross_product_sales", "amount": work.loc[work["IsGrossProductSale"], "LineAmount"].sum()},
        {"measure": "net_product_sales", "amount": work.loc[work["IsNetProductLine"], "LineAmount"].sum()},
        {"measure": "financial_total", "amount": work["LineAmount"].sum()},
    ]
)
display(sales_measure_summary)
```

### 4. 负价格、零价格、负数量要分开解释

价格异常不是一种问题，其中 `Price = 0` 且 `Quantity < 0` 的数字 Invoice 需要单独解释为商品损坏/库存损耗，而不是客户退货：

- `Price < 0` 只有 **5 行**，全部是 `A` 开头的 bad debt adjustment，不是商品负价。
- `Price = 0` 有 **6,202 行**，大量描述是 `lost`、`damaged`、`damages`、`wet`、`short`、`check` 或缺失描述，更像商品损坏、库存损耗、盘点差异和人工检查。
- `Quantity < 0` 不全是取消单：有 **3,457 行**是数字 Invoice、`Price = 0`、负数量，已标记为 `IsDamageOrInventoryLoss`。这些记录基本不影响销售金额，但会严重影响库存/销量分析。

因此清洗时不能只写 `Quantity < 0` 或 `Price <= 0` 删除，而要按交易语义分桶：客户退货走取消/退货口径；这批零价负数量记录走损坏/库存损耗口径。

```python
price_quantity_quality = pd.DataFrame(
    [
        {"issue": "Price < 0", "rows": int((work["Price"] < 0).sum())},
        {"issue": "Price = 0", "rows": int((work["Price"] == 0).sum())},
        {"issue": "Damage / inventory loss: Price = 0 and Quantity < 0", "rows": int(work["IsDamageOrInventoryLoss"].sum())},
    ]
)
display(price_quantity_quality)

display(
    work.loc[work["Price"] < 0, ["Invoice", "StockCode", "Description", "Quantity", "InvoiceDate", "Price", "Customer ID", "LineAmount"]]
)

display(
    work.loc[work["IsDamageOrInventoryLoss"], "Description"]
    .fillna("<missing>")
    .str.upper()
    .value_counts()
    .head(20)
    .rename_axis("Description")
    .reset_index(name="rows")
)
```

### 5b. Customer ID 缺失群体深挖

进一步检查后，`Customer ID` 缺失不能简单等同于商品损坏/库存损耗。更准确的结论是：**商品损坏/库存损耗是缺失 Customer ID 群体里的高置信子集，但缺失群体的主体仍是正常正向交易**。

关键证据：

- 缺失发生在 **invoice 级别**：没有一张 invoice 同时混合“有 Customer ID”和“缺 Customer ID”，说明这更像订单/渠道层面的客户绑定缺失，而不是单行字段随机坏掉。
- `Price = 0 & Quantity < 0` 的商品损坏/库存损耗共有 **3,457 行**，这些行 **100% 缺失 Customer ID**。
- 但这 3,457 行只占全部缺失 Customer ID 记录的 **1.42%**；缺失群体里约 **97.17%** 仍是 `Quantity > 0 & Price > 0` 的正常正向销售。
- 缺失群体 **98.77%** 来自 UK，并且 `DOTCOM POSTAGE` / `DOT` 高度集中在缺失群体，说明它可能包含匿名线上订单、dotcom/system channel、未绑定客户订单，以及库存损耗/系统调整。

因此，销售报表不能直接删除缺失 Customer ID 的记录；RFM 客户分析应排除或单独归入 `anonymous/system channel`；库存质量分析则应保留并单独追踪其中的损坏/库存损耗子集。


```python
# Customer ID 缺失深挖：它是订单/渠道层面的缺失，不是单纯商品损耗。
invoice_customer_completeness = (
    work.groupby("InvoiceStr")
    .agg(
        rows=("Invoice", "size"),
        missing_customer_rows=("IsCustomerMissing", "sum"),
        amount_sum=("LineAmount", "sum"),
        has_dotcom_postage=("StockCodeStr", lambda s: s.astype(str).str.upper().eq("DOT").any()),
        damage_loss_rows=("IsDamageOrInventoryLoss", "sum"),
    )
    .reset_index()
)
invoice_customer_completeness = invoice_customer_completeness.assign(
    customer_id_pattern=np.select(
        [
            invoice_customer_completeness["missing_customer_rows"].eq(0),
            invoice_customer_completeness["missing_customer_rows"].eq(invoice_customer_completeness["rows"]),
        ],
        ["all_present", "all_missing"],
        default="mixed_missing_present",
    )
)

display(
    invoice_customer_completeness.groupby("customer_id_pattern")
    .agg(
        invoices=("InvoiceStr", "size"),
        rows=("rows", "sum"),
        amount_sum=("amount_sum", "sum"),
        dotcom_postage_invoices=("has_dotcom_postage", "sum"),
        damage_loss_invoices=("damage_loss_rows", lambda s: int((s > 0).sum())),
    )
    .assign(
        invoice_pct=lambda x: (x["invoices"] / x["invoices"].sum() * 100).round(2),
        row_pct=lambda x: (x["rows"] / total_rows * 100).round(2),
    )
)

missing_customer = work.loc[work["IsCustomerMissing"]]
customer_missing_buckets = pd.DataFrame(
    [
        {
            "bucket": "normal positive line: Quantity > 0 and Price > 0",
            "rows": int((missing_customer["Quantity"].gt(0) & missing_customer["Price"].gt(0)).sum()),
            "amount_sum": missing_customer.loc[
                missing_customer["Quantity"].gt(0) & missing_customer["Price"].gt(0), "LineAmount"
            ].sum(),
        },
        {
            "bucket": "damage / inventory loss: Price = 0 and Quantity < 0",
            "rows": int(missing_customer["IsDamageOrInventoryLoss"].sum()),
            "amount_sum": missing_customer.loc[missing_customer["IsDamageOrInventoryLoss"], "LineAmount"].sum(),
        },
        {
            "bucket": "zero-price positive quantity",
            "rows": int((missing_customer["Price"].eq(0) & missing_customer["Quantity"].gt(0)).sum()),
            "amount_sum": 0.0,
        },
        {
            "bucket": "cancellation / return: C invoice",
            "rows": int(missing_customer["IsCancellation"].sum()),
            "amount_sum": missing_customer.loc[missing_customer["IsCancellation"], "LineAmount"].sum(),
        },
        {
            "bucket": "bad debt: A invoice",
            "rows": int(missing_customer["IsBadDebtAdjustment"].sum()),
            "amount_sum": missing_customer.loc[missing_customer["IsBadDebtAdjustment"], "LineAmount"].sum(),
        },
    ]
)
customer_missing_buckets = customer_missing_buckets.assign(
    pct_of_missing_rows=(customer_missing_buckets["rows"] / len(missing_customer) * 100).round(2),
    amount_sum=customer_missing_buckets["amount_sum"].round(2),
)
display(customer_missing_buckets)

display(
    missing_customer.groupby(["StockCodeStr", "Description"], dropna=False)
    .agg(rows=("Invoice", "size"), amount_sum=("LineAmount", "sum"), quantity_sum=("Quantity", "sum"))
    .sort_values("rows", ascending=False)
    .head(12)
    .reset_index()
)

print(
    "Damage/loss rows with missing Customer ID:",
    int((work["IsDamageOrInventoryLoss"] & work["IsCustomerMissing"]).sum()),
    "/",
    int(work["IsDamageOrInventoryLoss"].sum()),
)
print(
    "Damage/loss share within missing Customer ID rows:",
    pct((work["IsDamageOrInventoryLoss"] & work["IsCustomerMissing"]).sum(), work["IsCustomerMissing"].sum()),
    "%",
)
print(
    "Description missing rows with missing Customer ID:",
    int((work["IsDescriptionMissing"] & work["IsCustomerMissing"]).sum()),
    "/",
    int(work["IsDescriptionMissing"].sum()),
)

```

## Q1.1 小结：非显而易见的数据质量问题与处理策略

1. **Invoice 前缀有交易语义**：`C` 是取消/退货，`A` 是 bad debt adjustment。不要把所有行混在一个销售总额里；先定义 gross product sales、net product sales 和财务总额口径。
2. **特殊 StockCode 会污染商品分析**：邮费、折扣、manual adjustment、平台费用/佣金、礼品券、样品/测试/内部代码应单独分组。商品销量、热销 SKU、商品价格分析应排除这些非产品行；财务分析可以保留但要分项列示。
3. **极端 Quantity 有业务解释**：`80995` 和 `-80995` 是同一客户同一商品的销售与取消，间隔 12 分钟。它们不适合直接删除；做净销售时可互相抵消，做异常监控时应作为大额取消案例保留。
4. **负价格、零价格、负数量不是同一类问题**：负价格基本是 bad debt；零价格多是 lost/damaged/damages/wet/short/check/缺失描述；数字 Invoice 的负数量且零价格多为商品损坏或库存损耗。收入分析应排除这类零价损耗记录，库存质量分析应保留。
5. **Customer ID 缺失不是随机缺失**：缺失行占 22.77%，但金额占 13.68%，98.77% 来自 UK，平均行金额更低且高频出现 `DOTCOM POSTAGE`。客户分析要剔除或标记 anonymous/system channel；销售收入可保留但需单独披露。
6. **精确重复行不能无脑删除**：有 34,335 行精确重复，但缺少 invoice line id，无法判断是真重复还是同一订单同一商品被真实拆成多行。建议金额报表同时给出保留/去重敏感性，最终去重需业务主键支持。
7. **Description 混入库存/人工备注**：同一 StockCode 可能对应商品名和 `FOUND/MISSING/WRONGLY CODED/DAMAGED` 等备注。商品维表应优先使用 StockCode，并用最常见正常描述或外部商品表标准化名称。
8. **Description 缺失有明确模式**：缺失描述行全部零价格、全部缺 Customer ID，61.36% 为负数量。它们应从销售分析中排除，但可进入库存损耗/数据录入质量分析。
9. **2011-12 不是完整月份**：月度趋势、同比和季节性分析要排除或截断不完整月份，避免把 2011 年 12 月误判为销售下降。

**推荐清洗分层**：

- **商品销售 gross 口径**：保留数字 Invoice、产品型 StockCode、`Quantity > 0`、`Price > 0`；排除 `A/C`、非产品 StockCode、零价商品损坏/库存损耗、缺失 Description 的系统调整。
- **商品销售 net 口径**：在 gross 基础上纳入 `C` 取消/退货作为抵减，但仍排除 bad debt、平台费用、邮费、折扣、manual adjustment 等非商品行。
- **客户/RFM 口径**：仅使用有 Customer ID 的商品销售/退货记录；缺失 Customer ID 可作为 anonymous/system channel 单独统计，不混入客户生命周期指标。
- **库存/运营质量口径**：保留 `IsDamageOrInventoryLoss`、零价格、负数量、缺失 Description、FOUND/MISSING/DAMAGED/WRONGLY CODED 等记录，用于识别商品损坏、库存损耗、盘点和人工录入问题。
- **财务总额口径**：保留所有金额行，但必须把 product sales、returns、postage/carriage、discount、fees/commission、bad debt 分组展示。
